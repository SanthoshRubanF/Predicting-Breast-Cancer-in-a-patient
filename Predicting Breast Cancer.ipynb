# **Tuning the hyperparameters of SVM Classifier**

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
X = df.drop("diagnosis", axis=1).values # Fix: Include all relevant features
y = df.diagnosis

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
SC = StandardScaler()
X_train = SC.fit_transform(X_train)
X_test = SC.transform(X_test) # Fix: Only transform, do not fit on test data
```

```python
#Train the SVM classifier
SVM_classifier = SVC(probability=True)
SVM_classifier.fit(X_train, y_train)

#Predictions on the test set
y_pred = SVM_classifier.predict(X_test)

#Metrics
CM = confusion_matrix(y_test, y_pred)
TN, FP, FN, TP = CM.ravel()

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
specificity = TN / (TN + FP)
auc_roc = roc_auc_score(y_test, SVM_classifier.predict_proba(X_test)[:, 1])

print(f"--- SVM Performance ---")
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Sensitivity (Recall): {recall:.2f}")
print(f"Specificity: {specificity:.2f}")
print(f"AUC ROC: {auc_roc:.2f}")
```

```python
# Tuning the Hyperparameter with Grid Search
param_grid = {
    "C": [0.1, 1, 10, 100],
    "kernel": ["linear", "rbf", "poly"],
    "gamma": ["scale", "auto", 0.1, 1, 10],
}

# GridSearchCV
grid_search = GridSearchCV(estimator=SVC(probability=True), param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Best hyperparameters and the corresponding model
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

# Evaluation on test set
y_pred_best = best_model.predict(X_test)
TN_best, FP_best, FN_best, TP_best = confusion_matrix(y_test, y_pred_best).ravel()

print("Best hyperparameters:", best_params)
print(f"Accuracy on test set: {accuracy_score(y_test, y_pred_best):.2f}")
print(f"Specificity on test set: {TN_best / (TN_best + FP_best):.2f}")
```

```python
#Logistic Regression
logreg_model = LogisticRegression()
logreg_model.fit(X_train, y_train)
y_pred = logreg_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Logistic Regression Accuracy: {accuracy:.2f}")
```

```python
#Confusion Matrix
CM = confusion_matrix(y_test, grid_search.predict(X_test))

TP = CM[0][0]
TN = CM[1][1]
FN = CM[1][0]
FP = CM[0][1]

print(CM)
print(f"Confusion Matrix Testining Accuracy: {accuracy:.2f}")
```

# **Ensemble Techniques**

```python
# Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

TN_rf, FP_rf, FN_rf, TP_rf = confusion_matrix(y_test, y_pred_rf).ravel()

print("--- Random Forest Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}")
print(f"Precision: {precision_score(y_test, y_pred_rf):.2f}")
print(f"Sensitivity: {recall_score(y_test, y_pred_rf):.2f}")
print(f"Specificity: {TN_rf / (TN_rf + FP_rf):.2f}")
print(f"AUC ROC: {roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1]):.2f}")
```

```python
# AdaBoost Classifier
ada_model = AdaBoostClassifier(n_estimators=100, random_state=0)
ada_model.fit(X_train, y_train)
y_pred_ada = ada_model.predict(X_test)

TN_ada, FP_ada, FN_ada, TP_ada = confusion_matrix(y_test, y_pred_ada).ravel()

print("--- AdaBoost Performance ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_ada):.2f}")
print(f"Precision: {precision_score(y_test, y_pred_ada):.2f}")
print(f"Sensitivity: {recall_score(y_test, y_pred_ada):.2f}")
print(f"Specificity: {TN_ada / (TN_ada + FP_ada):.2f}")
print(f"AUC ROC: {roc_auc_score(y_test, ada_model.predict_proba(X_test)[:, 1]):.2f}")
```

# **Final Comparison of Models**

```python
models = {
    "Logistic Regression": logreg_model,
    "SVM (Tuned)": best_model,
    "Random Forest": rf_model,
    "AdaBoost": ada_model
}

results = []
for name, model in models.items():
    y_p = model.predict(X_test)
    tn, fp, fn, tp = confusion_matrix(y_test, y_p).ravel()
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_p),
        "Precision": precision_score(y_test, y_p),
        "Recall/Sens.": recall_score(y_test, y_p),
        "Specificity": tn / (tn + fp),
        "AUC ROC": roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])
    })

comparison_df = pd.DataFrame(results)
print(comparison_df.to_markdown())
```
